```python
input_file = open('Input.txt', 'r')
#input_file = open('Test.txt', 'r')
lines = input_file.readlines()

SIZE =  5

boards = []
numbers = lines[0].replace("\n","").split(',')
count = 0
row = 0
board = []
for line in lines:
    if count > 1 and len(line) > 1:
        print(line[0:2], line[3:5], line[6:8], line[9:11], line[12:14])
        board.append([int(line[0:2].replace(" ","")), int(line[3:5].replace(" ","")), int(line[6:8].replace(" ","")), int(line[9:11].replace(" ","")), int(line[12:14].replace(" ",""))])
        row = row + 1
        if row == 5:
            boards.append(board)
            board = []
            row = 0
    count = count + 1
#print(numbers, boards)

def mark_boards(value):
    for b in boards:
        for i in range(0, SIZE):
            for j in range(0, SIZE):
                if b[i][j] == value:
                    b[i][j] = -1

def sum_unmarked(b):
    result_sum = 0
    for i in range(0, SIZE):
            for j in range(0, SIZE):
                if b[i][j] != -1:
                    result_sum = result_sum + b[i][j]
    return result_sum
                    
def test_board(b):
    for i in range(0, SIZE):
        row = 0
        column = 0
        for j in range(0, SIZE):
            row = row + b[i][j]
            column = column + b[j][i]
        #print(row, column)
        if row == -5 or column == -5:
            return True
    return False

def test_all_board():
    for b in boards:
        if (test_board(b)):
            return sum_unmarked(b), b
    return 0, None

print(numbers, boards)
print("-----")
res = 0
final_b = []
final_n = 0
for n in numbers:
    mark_boards(int(n))
    print(boards)
    res, b = test_all_board()
    if (res > 0):
        final_res = res * int(n)
        final_b = b
        final_n = int(n)
        break
    print("-----")
print(final_res, res, final_n, final_b)
```
