```python
input_file = open('Input.txt', 'r')
#input_file = open('Test.txt', 'r')
lines = input_file.readlines()
```

```python
STEPS = 1000
flashes = 0
SIZE = 10
```

```python
def init_flashes():
    res = []
    for i in range(0, SIZE):
        tmp2 = []
        for j in range(0, SIZE):
            tmp2.append(False)
        res.append(tmp2)
    return res
```

```python
already_flashed = init_flashes()
```

```python
octopus = []
for line in lines:
    l = line.replace("\n", "")
    tmp = []
    for i in range(0, len(l)):
        tmp.append(int(l[i]))
    octopus.append(tmp)
print(octopus)
```

```python
def incr():
    for i in range(0, SIZE):
        for j in range(0, SIZE):
            octopus[i][j] = octopus[i][j] + 1
```

```python
def propagate(i, j):
    already_flashed[i][j] = True
    octopus[i][j] = 0
    global flashes
    flashes = flashes + 1
    for ii in [-1, 0, 1]:
        for jj in [-1, 0, 1]:
            new_i = i+ii
            new_j = j+jj
            if (new_i >=0 and new_i <=9 and new_j>= 0 and new_j<=9):                    
                if (already_flashed[new_i][new_j] == False):
                    #print("Propagate", i, j, new_i, new_j)
                    octopus[new_i][new_j] = octopus[new_i][new_j] + 1
                    if (octopus[new_i][new_j] > 9):
                        propagate(new_i,new_j)
```

```python
def check_flashes():
    for i in range(0, SIZE):
        for j in range(0, SIZE):
            if (octopus[i][j] > 9):
                propagate(i,j)
```

```python
def check():
    res = 0
    for i in range(0, SIZE):
        for j in range(0, SIZE):
            res = res + octopus[i][j]            
    return res
```

```python
print(octopus)
for i in range(0, STEPS):
    incr()
    check_flashes()
    print(i+1, flashes, octopus)
    already_flashed = init_flashes()
    if (check() == 0):
        print("SYNC", i+1)
        break
print(flashes)
```
