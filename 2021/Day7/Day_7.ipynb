```python
input_file = open('Input.txt', 'r')
#nput_file = open('Test.txt', 'r')
lines = input_file.readlines()
```

```python
crabs = lines[0].replace("\n","").split(',')
max_pos = -1
for i in range(0, len(crabs)):
    crabs[i] = int(crabs[i])
    if (crabs[i] > max_pos):
        max_pos = crabs[i]
print(max_pos)
```

```python
def fact(value):
    factorial = 0
    for i in range (0,value+1):
        factorial = factorial + i
    return factorial
```

```python
def compute(pos):
    result = 0
    for i in range(0, len(crabs)):
        result = result + abs(crabs[i] - pos)
    return result
```

```python
def compute2(pos):
    result = 0
    for i in range(0, len(crabs)):
        result = result + fact(abs(crabs[i] - pos))
    return result
```

```python
min_fuel = 999999
min_pos = -1
for i in range(0, max_pos+1):
    res = compute(i)
    if (res < min_fuel):
        min_fuel = res
        min_pos = i
    #print(i, res)
print(min_fuel, min_pos)
```
