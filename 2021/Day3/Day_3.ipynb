```python
input_file = open('Input.txt', 'r')
#input_file = open('Test.txt', 'r')
lines = input_file.readlines()

SIZE =  12 #5 #12

rows = []
zeros = [0 for x in range(0, SIZE)]
ones = [0 for x in range(0, SIZE)]
for line in lines:
    line = line.replace("\n","")
    for i in range(0, SIZE):
        value = int(line[i])
        if value == 1:
            ones[i] = ones[i]+ 1
        else:
            zeros[i] = zeros[i]+ 1
    rows.append(line)
    #print(line, zeros, ones)            
print(zeros, ones)

gamma = ""
epsilon = ""
for i in range(0, SIZE):
    if (zeros[i] > ones[i]):
        gamma = gamma + "0"
        epsilon = epsilon + "1"
    else:
        gamma = gamma + "1"
        epsilon = epsilon + "0"
print(gamma, epsilon)
print(int(gamma, 2), int(epsilon, 2))
print(int(gamma, 2) * int(epsilon, 2))
```

```python
def return_result(lines, preference):
    rows = []
    zeros = [0 for x in range(0, SIZE)]
    ones = [0 for x in range(0, SIZE)]
    for line in lines:
        line = line.replace("\n","")
        for i in range(0, SIZE):
            value = int(line[i])
            if value == 1:
                ones[i] = ones[i]+ 1
            else:
                zeros[i] = zeros[i]+ 1
        rows.append(line)
        #print(line, zeros, ones)            

    gamma = ""
    epsilon = ""
    for i in range(0, SIZE):
        if (zeros[i] == ones[i]):
            #print("PREF")
            if (preference == "OXY"):
                gamma = gamma + "1"
                epsilon = epsilon + "1"
            else:
                gamma = gamma + "0"
                epsilon = epsilon + "0"
        else:                
            if (zeros[i] > ones[i]):
                gamma = gamma + "0"
                epsilon = epsilon + "1"
            else:
                gamma = gamma + "1"
                epsilon = epsilon + "0"
    return gamma, epsilon
```

```python
def eliminate(data, pos, value):
    print("Keep " + str(value) + " at pos " + str(pos+1))
    result = []
    if (len(data) > 1):
        for d in data:
            if (d[pos] == value):
                result.append(d)
    else:
        result = data
    return result 
```

```python
oxygen = rows
co2 = rows
#print(rows)
for i in range (0, SIZE): 
    print("Round " + str(i+1))
    gamma, epsilon = return_result(oxygen, "OXY")
    print(gamma)    
    oxygen = eliminate(oxygen, i, gamma[i])    
    print(oxygen)
    
    gamma_co2, epsilon_co2 = return_result(co2, "CO2")
    print(epsilon_co2)    
    co2 = eliminate(co2, i, epsilon_co2[i])    
    print(co2)
    
print(int(oxygen[0],2), int(co2[0],2))
print(int(oxygen[0],2) * int(co2[0],2))
```
