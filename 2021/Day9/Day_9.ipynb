```python
input_file = open('Input.txt', 'r')
#input_file = open('Test.txt', 'r')
lines = input_file.readlines()
```

```python
rows = len(lines)
columns = len(lines[0])-1
table = []
print(rows, columns)
for i in range(0, rows):
    tmp = []
    l = lines[i].replace("\n","")
    for j in range(0, columns):
        tmp.append(int(lines[i][j]))
    table.append(tmp)
print(table)
```

```python
display = ""
for i in range(0, rows):
    for j in range(0, columns):
        display = display + str(table[i][j])
    display = display + "\n"
print(display)
```

```python
def test_low(table, i, j):
    ref = table[i][j]
    low = True
    #print(i, j, ref)
    if i > 0:
        comp = table[i-1][j]
        #print("Above", i-1, j, comp)
        if (comp <= ref):
            low = False
    if i < rows-1:
        comp = table[i+1][j]
        #print("Below", i+1, j, comp)
        if (comp <= ref):
            low = False
    if j > 0:
        comp = table[i][j-1]
        #print("Left ", i, j-1, comp)
        if (comp <= ref):
            low = False
    if j < columns-1:
        comp = table[i][j+1]
        #print("Right", i, j+1, comp)
        if (comp <= ref):
            low = False
    #print("Res", low)
    return low
```

```python
lows = []
for i in range(0, rows):
    tmp = []
    for j in range(0, columns):
        tmp.append(test_low(table, i, j))
    lows.append(tmp)
print(lows)
```

```python
res = 0
for i in range(0, rows):
    for j in range(0, columns):
        if (lows[i][j] == True):
            res = res + (1+table[i][j])
print(res) # 15 # 417
```

```python
basins = []
for i in range(0, rows):
    tmp = []
    for j in range(0, columns):
        tmp.append(False)
    basins.append(tmp)
print(basins)
```

```python
def explore(i, j):
    res = 1    
    if i > 0:
        if table[i-1][j] != 9 and basins[i-1][j] == False:
            basins[i-1][j] = True
            res = res + explore(i-1, j)
    if i < rows-1:
        if table[i+1][j] != 9 and basins[i+1][j] == False:
            basins[i+1][j] = True
            res = res + explore(i+1, j)
    if j > 0:
        if table[i][j-1] != 9 and basins[i][j-1] == False:
            basins[i][j-1] = True
            res = res + explore(i, j-1)
    if j < columns-1:
        if table[i][j+1] != 9 and basins[i][j+1] == False:
            basins[i][j+1] = True
            res = res + explore(i, j+1)    
    return res
```

```python
all_basins = []
for i in range(0, rows):
    for j in range(0, columns):
        if lows[i][j] == True:
            expl = explore(i, j)-1
            #print(i, j, expl)
            all_basins.append(expl)
all_sorted = sorted(all_basins)
print(all_sorted)
print(all_sorted[-1]* all_sorted[-2]* all_sorted[-3] )
```
