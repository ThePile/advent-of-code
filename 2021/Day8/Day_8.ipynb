```python
def convert(decoding, number):
    code = []
    if number == 0:
        code = ["t", "tl", "tr", "bl", "br", "b"]
    elif number == 1:
        code = ["tr", "br"]
    elif number == 2:
        code = ["t", "tr", "m", "bl", "b"]
    elif number == 3:
        code = ["t", "tr", "m", "br", "b"]
    elif number == 4:
        code = ["tl", "tr", "m", "br"]
    elif number == 5:
        code = ["t", "tl", "m", "br", "b"]
    elif number == 6:
        code = ["t", "tl", "m", "bl", "br", "b"]
    elif number == 7:
        code = ["t", "tr", "br"]
    elif number == 8:
        code = ["t", "tl", "tr", "m", "bl", "br", "b"]
    elif number == 9:
        code = ["t", "tl", "tr", "m", "br", "b"]
    result = ""
    for c in code:
        result = result + decoding[c]
    result = "".join(sorted(result))
    return result
```

```python
def analyze(sizes):
    all = "abcdefg"
    decoding_val = {}
    decoding_pos = {}
    for i in ['t', 'tl','tr','m','bl', 'br','b']:
        decoding_pos[i] = all
    for i in range(0, 10):
        decoding_val[i] = all
    decoding_val[1] = sizes[2][0]
    decoding_val[7] = sizes[3][0]
    decoding_val[4] = sizes[4][0]
    decoding_val[8] = sizes[7][0]

    if (2 in sizes and 3 in sizes):
        tmp = sizes[3][0]
        for l in sizes[2][0]:
            tmp= tmp.replace(l, "")
        decoding_pos["t"] = tmp
        for d in decoding_pos:
            if (d != "t"):
                decoding_pos[d] = decoding_pos[d].replace(tmp, "")
    if (2 in sizes):
        for l in sizes[2][0]:
            decoding_pos['t'] = decoding_pos['t'].replace(l, "")
            decoding_pos['tl'] = decoding_pos['tl'].replace(l, "")
            decoding_pos['m'] = decoding_pos['m'].replace(l, "")
            decoding_pos['bl'] = decoding_pos['bl'].replace(l, "")
            decoding_pos['b'] = decoding_pos['b'].replace(l, "")
    if (2 in sizes and 4 in sizes):
        tmp = sizes[4][0]
        for l in sizes[2][0]:
            tmp= tmp.replace(l, "")
        for l in tmp:
            decoding_pos['t'] = decoding_pos['t'].replace(l, "")
            decoding_pos['tr'] = decoding_pos['tr'].replace(l, "")
            decoding_pos['br'] = decoding_pos['br'].replace(l, "")
            decoding_pos['bl'] = decoding_pos['bl'].replace(l, "")
            decoding_pos['b'] = decoding_pos['b'].replace(l, "")                
    if (0 in sizes and 7 in sizes):
        tmp = sizes[7][0]
        for l in sizes[7][0]:
            tmp= tmp.replace(l, "")
        decoding_pos["t"] = tmp
    
    if (decoding_pos["b"] == decoding_pos["bl"]):
        for l in decoding_pos["b"]:
            decoding_pos['t'] = decoding_pos['t'].replace(l, "")
            decoding_pos['tr'] = decoding_pos['tr'].replace(l, "")
            decoding_pos['tl'] = decoding_pos['tl'].replace(l, "")
            decoding_pos['m'] = decoding_pos['m'].replace(l, "")
            decoding_pos['br'] = decoding_pos['br'].replace(l, "")     
    if (decoding_pos["tl"] == decoding_pos["m"]):
        for l in decoding_pos["m"]:
            decoding_pos['t'] = decoding_pos['t'].replace(l, "")
            decoding_pos['tr'] = decoding_pos['tr'].replace(l, "")
            decoding_pos['br'] = decoding_pos['br'].replace(l, "")
            decoding_pos['b'] = decoding_pos['b'].replace(l, "")
            decoding_pos['bl'] = decoding_pos['bl'].replace(l, "")   


    for l in sizes[4]:
        for ll in sizes[6]:
            if l[0] in ll and l[1] in ll and l[2] in ll and l[3] in ll:
                decoding_val[9] = ll
                for a in all:
                    if a not in ll:
                        decoding_pos['bl'] = a 
                        decoding_pos['b'] =  decoding_pos['b'].replace(a, "")  
                        decoding_val[3] = decoding_val[3].replace(a, "")  
                        decoding_val[5] = decoding_val[5].replace(a, "") 
    
    for l in sizes[3]:
        for ll in sizes[5]:
            if l[0] in ll and l[1] in ll and l[2] in ll:
                decoding_val[3] = ll
                ll_tmp = ll
                for a in l:
                    ll_tmp = ll_tmp.replace(a, "")
                decoding_pos['m'] = ll_tmp.replace(decoding_pos['b'], "")   
                #decoding_pos['m'] = decoding_pos['m'].replace(a, "")
                decoding_pos['tl'] = decoding_pos['tl'].replace(decoding_pos['m'], "")
                for ii in [0, 1, 7]:
                    decoding_val[ii] = decoding_val[ii].replace(decoding_pos['m'], "") 
                for ii in [2,3,7]:
                    decoding_val[ii] = decoding_val[ii].replace(decoding_pos['tl'], "")        
    
    for aa in sizes[5]:
        if decoding_pos['tr'][0] in aa and decoding_pos['tr'][1] in aa:
            decoding_val[3] = aa
        
        
    size6 = []
    for ss in sizes[6]:
        size6.append(ss)
    for aa in sizes[6]:
        if decoding_pos['tr'][0] in aa and decoding_pos['tr'][1] in aa:
            size6.remove(aa)
          
            
    decoding_val[6] = size6[0]
    
    tmp_tr = decoding_pos['tr']
    for a in tmp_tr:
        if a not in decoding_val[6]:
            decoding_pos['tr'] = a
            decoding_pos['br'] = decoding_pos['br'].replace(a, "")
        
    for i in range(0,10):
        decoding_val[i] = "".join(sorted(decoding_val[i]))
    print("POS", decoding_pos)
    print("VALUE", decoding_val)
    #print("---")
    return decoding_pos

ALLRES = 0
for l in lines:
    sizes = {}
    right = l.replace("\n","").split(" | ")[0]
    left = l.replace("\n","").split(" | ")[1]
    print(right, " | ", left)
    all_right = right.split(" ")
    all_left_non = left.split(" ")
    all_left = []
    for l in all_left_non:
        all_left.append("".join(sorted(l)))
    for r in all_right:
        s = len(r)
        rr = "".join(sorted(r))
        if s in sizes:
            sizes[s].append(rr)
        else:
            sizes[s] = [rr]
    print(sizes)
    decoding = analyze(sizes)
    compute = {}
    for i in range(0, 10):
        kk = convert(decoding, i)
        kk = "".join(sorted(kk))
        compute[kk]= i
    res = ""
    print(compute)
    for l in all_left:
        lll = compute[l]
        #print(l, lll)
        res = res + str(lll)
    print(res)
    ALLRES = ALLRES + int(res)
print(ALLRES)
```
