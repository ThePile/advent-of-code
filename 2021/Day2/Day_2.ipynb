```python
input_file = open('Input.txt', 'r')
lines = input_file.readlines()
 
count = 0
hor = 0
dep = 0
aim = 0
for line in lines:
    line = line.replace("\n","")
    actions = line.split(" ")
    if (actions[0] == 'forward'):
        hor = hor + int(actions[1])
        dep = dep + (aim * int(actions[1]))
    else:
        if (actions[0] == 'up'):
            aim = aim - int(actions[1])
        else: # down
            aim = aim + int(actions[1])
    print(line, actions, hor, dep, aim)
print(dep, hor, aim, dep*hor)
```
