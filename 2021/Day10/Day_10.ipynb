```python
input_file = open('Input.txt', 'r')
#input_file = open('Test.txt', 'r')
lines = input_file.readlines()
```

```python
scoring = { ")": 3, "]": 57, "}": 1197, ">": 25137 }
```

```python
opening = ["(", "[", "{", "<" ]
closing = [")", "]", "}", ">" ]
matching = { ")": "(" , "]":"[", "}":"{", ">":"<" }
reverse_matching = { "(":")" , "[":"]", "{":"}", "<":">" }
keep = []
without = []
for line_tmp in lines:
    buffer = []
    line = line_tmp.replace("\n", "")
    error = False
    expected = ""
    current = ""
    pos = -1
    for i in range(0, len(line)):
        l = line[i]
        if (l in opening):
            buffer.append(l)
        else:
            if (buffer[-1] == matching[l]):
                buffer.pop()
            else:                
                expected = reverse_matching[buffer[-1]]
                current = l
                pos = i
                error = True
                break
    if (error == True):
        #print(line, "  ", pos, " Expected ", expected, " but found ", current)
        keep.append(current)
    else:
        without.append(line)
print(keep)
```

```python
res = 0
for k in keep:
    res = res + scoring[k]
print(res)
```

```python
scores = []
for w in without:    
    buffer = []
    for i in range(0, len(w)):
        l = w[i]
        if (l in opening):
            buffer.append(l)
        else:
            if (buffer[-1] == matching[l]):
                buffer.pop()
    #print(buffer)
    result = []
    for b in buffer:
        result.insert(0, reverse_matching[b])
    res = 0    
    for kk in result:
        res = res*5 + scoring2[kk]
    scores.append(res)
    print(w, result, res)
scores = sorted(scores)
print(scores)
print(scores[int(len(scores)/2)])
```
